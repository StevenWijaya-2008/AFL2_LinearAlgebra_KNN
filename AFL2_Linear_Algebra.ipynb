```python
import numpy as np
import pandas as pd

df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Data_KNN_LinearAlgebra.xlsx")
data = df.values
print("="*20, "DATA AWAL", "="*20)
print(data)
print("\n")

# TEST CASE
target = np.array([31, 6, 5, 16, 4])
print("===== TEST CASE =====")
print(target)
print("\n")

# Menambah kolom kosong untuk menyimpan hasil perhitungan jarak
new_col = np.zeros((data.shape[0],1))
data = np.hstack((data, new_col))

# Menghitung jarak
for i in data:
  d = np.linalg.norm(i[:-2] - target)
  i[-1] =  d

k = 5
data = data[np.argsort(data[:, -1])] # SORT BERDASARKAN JARAK

# MENGHITUNG W
new_col = np.zeros((data.shape[0],1))
data = np.hstack((data, new_col))
data[:, -1] = 1 / ((data[:, -2] + 0.00001)**2)

print("="*20, "DATA TERDEKAT (K = 5)", "="*20)
for row in data[:k]: #UNTUK MENAMPILKAN 3 ANGKA DI BELAKANG KOMA
    formatted_row = [
        "{:.3f}".format(val) if isinstance(val, (float, np.float64)) else str(val)
        for val in row
    ]
    print(formatted_row)
print("\n")

answers = data[:,-3][:k]
uniq, count = np.unique(answers, return_counts=True)
print("==== HASIL PREDIKSI K = 5 (TANPA WEIGHTED) ====")
print(uniq[np.argmax(count)])
print("\n")

# MENGHITUNG BERDASARKAN WEIGHTED
skor_total = {}
for row in data[:k]:
    label = row[-3]
    weight = row[-1]
    if label in skor_total:
        skor_total[label] += weight
    else:
        skor_total[label] = weight

# TAMPILKAN SKOR
print("===== TOTAL WEIGHT PER KATEGORI =====")
for kategori, total_w in skor_total.items():
    print(f"{kategori}: {total_w:.4f}")

print("\n")
pemenang = max(skor_total, key=skor_total.get)
print("===== HASIL PREDIKSI K = 5 (WEIGHTED) =====")
print(f"Pemenangnya adalah: {pemenang}")
```

```python
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Data_KNN_LinearAlgebra.xlsx")
data = df.values
print("="*20, "DATA AWAL", "="*20)
print(data)
print("\n")

# NORMALISASI
X = data[:, :-1]  # Angka
y = data[:, -1]   # Segment
X = X.astype(float) # Memastikan X tipe data float
X_norm = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
y = y.reshape(-1, 1) # y jadi bentuk kolom
data_normalized = np.hstack((X_norm, y))
np.set_printoptions(precision=3, suppress=True)

print("="*20, "DATA SETELAH NORMALISASI", "="*20)
for row in data_normalized:
    # format 3 angka di belakang koma
    formatted_row = ["{:.3f}".format(val) if isinstance(val, (int, float, np.float64)) else str(val) for val in row]
    print(formatted_row)
print("\n")


# TEST CASE
target = np.array([39, 9, 9, 25, 5])
target_norm = (target - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
print("===== TEST CASE =====")
print(target_norm)
print("\n")


# Menambah kolom kosong untuk menyimpan hasil perhitungan jarak
new_col = np.zeros((data_normalized.shape[0],1))
data_normalized = np.hstack((data_normalized, new_col))

# Menghitung jarak
for i in data_normalized:
  d = np.linalg.norm(i[:-2] - target_norm)
  i[-1] =  d

k = 5
data_normalized = data_normalized[np.argsort(data_normalized[:, -1])] # SORT BERDASARKAN JARAK

# MENGHITUNG W
new_col = np.zeros((data_normalized.shape[0],1))
data_normalized = np.hstack((data_normalized, new_col))
data_normalized[:, -1] = 1 / ((data_normalized[:, -2] + 0.00001)**2)

print("="*20, "DATA TERDEKAT (K = 5)", "="*20)
for row in data_normalized[:k]: #UNTUK MENAMPILKAN 3 ANGKA DI BELAKANG KOMA
    formatted_row = [
        "{:.3f}".format(val) if isinstance(val, (float, np.float64)) else str(val)
        for val in row
    ]
    print(formatted_row)
print("\n")


answers = data_normalized[:,-3][:k]
uniq, count = np.unique(answers, return_counts=True)
print("==== HASIL PREDIKSI K = 5 (TANPA WEIGHTED) ====")
print(uniq[np.argmax(count)])
print("\n")

# MENGHITUNG BERDASARKAN WEIGHTED
skor_total = {}
for row in data_normalized[:k]:
    label = row[-3]
    weight = row[-1]
    if label in skor_total:
        skor_total[label] += weight
    else:
        skor_total[label] = weight

# TAMPILKAN SKOR
print("===== TOTAL WEIGHT PER KATEGORI =====")
for kategori, total_w in skor_total.items():
    print(f"{kategori}: {total_w:.4f}")

print("\n")
pemenang = max(skor_total, key=skor_total.get)
print("===== HASIL PREDIKSI K = 5 (WEIGHTED) =====")
print(f"Pemenangnya adalah: {pemenang}")
```
